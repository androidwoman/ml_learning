```python
import pandas as pd
```

```python
# Load the 'books' dataset from the CSV file
books_dataset = pd.read_csv('books.csv')

# Load the 'authors' dataset from the CSV file
authors_dataset = pd.read_csv('authors.csv')

# Load the 'language' dataset from the CSV file
language_dataset = pd.read_csv('language.csv')

# Load the 'publisher' dataset from the CSV file
publisher_dataset = pd.read_csv('publisher.csv')

# Print the 'books' dataset
print("books_dataset:\n", books_dataset)

# Print the 'authors' dataset
print("\nauthors_dataset:\n", authors_dataset)

# Print the 'language' dataset
print("\nlanguage_dataset:\n", language_dataset)

# Print the 'publisher' dataset
print("\npublisher_dataset:\n", publisher_dataset)
```

```python
# Adding 'author_id' column to df_authors
authors_dataset = authors_dataset.assign(author_id=authors_dataset.index)

# Printing the updated df_authors
print("Updated df_authors:\n", authors_dataset)

# Adding 'lang_id' column to df_language
language_dataset = language_dataset.assign(lang_id=language_dataset.index)

# Printing the updated df_language
print("Updated df_language:\n", language_dataset)

# Adding 'publisher_id' column to df_publisher
publisher_dataset = publisher_dataset.assign(publisher_id=publisher_dataset.index)

# Printing the updated df_publisher
print("Updated df_publisher:\n", publisher_dataset)
```

```python

books_dataset = pd.merge(books_dataset, authors_dataset, on='author_id')


books_dataset = pd.merge(books_dataset, language_dataset, on='lang_id')


books_dataset = pd.merge(books_dataset, publisher_dataset, on='publisher_id')


print("Updated book:\n", books_dataset)
```

```python

books_dataset = books_dataset.drop('author_id', axis=1)
books_dataset = books_dataset.drop('lang_id', axis=1)
books_dataset = books_dataset.drop('publisher_id', axis=1)

print("new book: db\n",books_dataset)
```

```python

books_dataset = books_dataset.rename(columns={"name_x": "author", "code": "lang", "name_y": "publisher"})
books_dataset.head()
```

```python

c = 0 
sum= 0
publisher_dataset['publisher_average'] = ""
for index1, value1 in enumerate(publisher_dataset['name']):
  for index2, value2 in enumerate(books_dataset['publisher']):
    if value1 == value2:
      c += 1
      sum += books_dataset['average_rating'].iloc[index2]
  publisher_average_rating = sum / c
  publisher_dataset['publisher_average'].iloc[index1] = publisher_average_rating
  c =0
  sum = 0
  publisher_average_rating = 0
print("publisher:",publisher_dataset)
```

```python
publisher_dataset = publisher_dataset.sort_values(by=['publisher_average'], ascending=False) 
print(publisher_dataset)
```

```python
best = 0
best_publisher = []
for index, value in enumerate(publisher_dataset['publisher_average']):
  if value == 5.0:
    best_publisher.append(publisher_dataset['name'].iloc[index])
    best +=1
print("best publishers:\n",best_publisher)
print("\nbest:\t",best)
```
