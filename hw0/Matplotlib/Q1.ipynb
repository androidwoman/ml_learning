```python
import zipfile
import matplotlib.pyplot as plt
import io
import numpy as np
```

```python


# Open the zip file and read the logo image
zip_file_path = "logo.zip"  # Path to the zip file
logo_file_name = "logo.npz"  # Name of the logo image file inside the zip file

with zipfile.ZipFile(zip_file_path, "r") as zip_file:
    with zip_file.open(logo_file_name) as logo_file:
        logo_data = logo_file.read()

logo_data = np.load(io.BytesIO(logo_data))

# Extract the logo image from the data
logo_image = logo_data['arr_0']

# Display the image
plt.imshow(logo_image)
plt.title("University of Tehran Logo")
plt.axis('off')  # Remove x and y axes from the display
plt.show()

# Report the image dimensions
height, width, _ = logo_image.shape
print("Image width:", width)
print("Image height:", height)
```
