```python
import matplotlib.pyplot as plt

# Define the percentages for each slice
percentages = [4.8] * 19 + [9.5]  # 19 slices with 4.8% and 1 slice with 9.5%

# Labels for each slice
labels = ['Slice ' + str(i+1) for i in range(len(percentages))]

# Create the pie chart
plt.figure(figsize=(8, 8))
plt.pie(percentages, labels=labels, autopct='%1.1f%%', startangle=140)

# Add a title
plt.title('Pie Chart with 20 Slices')

plt.show()
```
