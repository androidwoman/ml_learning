```python
import matplotlib.pyplot as plt

x = range(12)
negative_data = [-0.51, -0.79, -0.47, -0.43, -0.37, -0.44, -0.29, -0.39, -0.33, -0.17, -0.14, -0.07]
positive_data = [0.63, 0.72, 0.77, 0.55, 0.60, 0.34, 0.41, 0.27, 0.17, 0.16, 0.13, 0.06]

fig = plt.figure()
ax = plt.subplot(111)
# Plot positive data in pink
ax.bar(x, positive_data, width=1, color='skyblue', label='Positive Data')
for i, val in enumerate(negative_data):
    ax.text(x[i], val, str(val), ha='center', va='bottom')
for i, val in enumerate(positive_data):
    ax.text(x[i], val, str(val), ha='center', va='top')

# Plot negative data in sky blue
ax.bar(x, negative_data, width=1, color='pink', label='Negative Data')


# Add legend
ax.legend()

plt.show()
```
