```python
import matplotlib.pyplot as plt


# Create a figure
fig = plt.figure()

# Define the coordinates and size for each subplot
subplots = [
    [0.1, 0.1, 0.5, 0.5],
    [0.2, 0.2, 0.5, 0.5],
    [0.3, 0.3, 0.5, 0.5],
    [0.4, 0.4, 0.5, 0.5]
]

# Create and plot on each nested subplot
for subplot_coords in subplots:
    ax = fig.add_axes(subplot_coords)
    ax.plot()
    ax.set_title('Nested Subplot')

# Show the nested subplots
plt.show()
```
