```python
import numpy as np
import matplotlib.pyplot as plt


data = np.loadtxt("populations.txt", skiprows=1, usecols=(0, 1, 2, 3), dtype=float)

# Extract year and population data for each species
years = data[:, 0][:50]
hare_population = data[:, 1][:50]
lynx_population = data[:, 2][:50]
carrot_population = data[:, 3][:50]
print(years)

# Define the "unwanted years"
unwanted_years = [1900, 1901, 1902, 1903, 1904, 1913, 1915, 1922, 1924, 1945, 1946]

# Filter 
valid_years = [year for year in years if year not in unwanted_years]
valid_data = data[:50][~np.isin(years, unwanted_years)]

# Calculate mean and standard deviation

years = valid_years
hare_population = valid_data[:,1]
lynx_population = valid_data[:,2]
carrot_population = valid_data[:,3]

mean_hare = np.mean(hare_population)
std_hare = np.std(hare_population)
mean_lynx = np.mean(lynx_population)
std_lynx = np.std(lynx_population)
mean_carrot = np.mean(carrot_population)
std_carrot = np.std(carrot_population)




# Create a plot to visualize the population trend for the first 50 years
plt.figure(figsize=(10, 6))
plt.plot(valid_years, hare_population, label='Hare Population',marker='o')
plt.plot(valid_years, lynx_population, label='Lynx Population',marker='o')
plt.plot(valid_years, carrot_population, label='Carrot Population',marker='o')
plt.xlabel('Years')
plt.ylabel('Population')
plt.title('Population Trend for Hares, Lynxes, and Carrots (1900-1949)')
plt.legend()
plt.grid(True)
plt.show()

# Print the mean and standard deviation of population for each species
print("Mean Hare Population:", mean_hare)
print("Standard Deviation Hare Population:", std_hare)
print("Mean Lynx Population:", mean_lynx)
print("Standard Deviation Lynx Population:", std_lynx)
print("Mean Carrot Population:", mean_carrot)
print("Standard Deviation Carrot Population:", std_carrot)
```
