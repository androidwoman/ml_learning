# calculate distance

```python
import timeit


def calculate_distance_loops():
    distances = []
    for new_point in new_irises:
        distance_list = []
        for data_point in irises_data:
            distance = np.sqrt(np.sum((data_point - new_point) ** 2))
            message = f"Distance of {new_point} from {data_point} is: {distance}"
            distance_list.append(message)
        distances.append(distance_list)
        
    print("calculate_distance_loops",distances)        


def calculate_distance_single_loop():
    distances = []
    for new_point in new_irises:
        distance_list = [np.sqrt(np.sum((data_point - new_point) ** 2)) for data_point in irises_data]
        distances.append(distance_list)
    print("calculate_distance_single_loop" ,distances)   


def calculate_distance_vectorized():
    distances = np.linalg.norm(irises_data[:, np.newaxis, :] - new_irises, axis=2)
    print("calculate_distance_vectorized",distances)

loops_time = timeit.timeit(calculate_distance_loops,number=1 )
single_loop_time = timeit.timeit(calculate_distance_single_loop,number=1 )
vectorized_time = timeit.timeit(calculate_distance_vectorized,number=1 )

print("Execution time with two loops :", loops_time)
print("Execution time with a single loop :", single_loop_time)
print("Execution time with vectorization :", vectorized_time)


```
