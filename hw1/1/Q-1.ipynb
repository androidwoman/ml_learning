```python


np.random.seed(42)

# Shuffle the rows of the DataFrame
df_shuffled = df.sample(frac=1)

# Calculate the number of rows for training and test sets
num_samples = len(df_shuffled)
train_size = int(0.8 * num_samples)

# Split the data into training and test sets
train_data = df_shuffled[:train_size]
test_data = df_shuffled[train_size:]

# Separate the labels (target) from the data
train_labels = train_data['Attrition']
test_labels = test_data['Attrition']

# Remove the 'Attrition' column from the data
train_data = train_data.drop(columns=['Attrition'])
test_data = test_data.drop(columns=['Attrition'])
```

```python
class OneR(object):
    
    def __init__(self):
        self.ideal_variable = None
        self.rule=None
        self.max_accuracy = 0
    
    def fit(self, X, y):
        response = list()
        result = dict()
        
        dfx = pd.DataFrame(X)
        
        for i in dfx:
            result[str(i)] = dict()
            join_data = pd.DataFrame({"v":dfx[i], "c":y})
            cross_table = pd.crosstab(join_data.v, join_data.c)
            summary = cross_table.idxmax(axis=1)
            result[str(i)] = dict(summary)
            
            counts = 0
            
            for idx, row in join_data.iterrows():
                if row['c'] == result[str(i)][row['v']]:
                    counts += 1

            accuracy = (counts/len(y))
            
            if accuracy > self.max_accuracy:
                self.max_accuracy = accuracy
                self.ideal_variable = i
                self.rule=result[str(i)]

            result_feature = {"variable": str(i), "accuracy":accuracy, "rules": result[str(i)] }  
            response.append(result_feature)
            
            
        print(response)   
        print("best rule is for feature:\n",str(self.ideal_variable) ,"\n rule is : \n" ,self.rule ) 
        return response

    
    def predict(self, X=None):
        dfx = pd.DataFrame(X)
        predicted_labels = []
        for idx, row in dfx.iterrows():
            predicted_label = self.rule.get(row[self.ideal_variable], None)
            predicted_labels.append(predicted_label)

        return predicted_labels
        
    
           
       
```

```python
rule = OneR()
results = rule.fit(train_data, train_labels)
```

```python
predicted_label = rule.predict(test_data,)
print("predict classes : " ,predicted_label)
```

```python
def confusion_matrix(true_labels, predicted_labels, positive_label):
    true_positive = 0
    true_negative = 0
    false_positive = 0
    false_negative = 0

    for true, pred in zip(true_labels, predicted_labels):
        if true == positive_label and pred == positive_label:
            true_positive += 1
        elif true != positive_label and pred != positive_label:
            true_negative += 1
        elif true != positive_label and pred == positive_label:
            false_positive += 1
        elif true == positive_label and pred != positive_label:
            false_negative += 1

    return {
        "TP": true_positive,
        "TN": true_negative,
        "FP": false_positive,
        "FN": false_negative
    }

def precision(confusion_matrix):
    true_positive = confusion_matrix["TP"]
    false_positive = confusion_matrix["FP"]
    return true_positive / (true_positive + false_positive)

def recall(confusion_matrix):
    true_positive = confusion_matrix["TP"]
    false_negative = confusion_matrix["FN"]
    return true_positive / (true_positive + false_negative)

def f1_score(precision, recall):
    return 2 * (precision * recall) / (precision + recall)

def roc_auc_score(true_labels, predicted_labels):
    num_positive = sum(1 for label in true_labels if label == positive_label)
   

    true_positive_rate = sum(1 for true, pred in zip(true_labels, predicted_labels) if true == positive_label and pred == positive_label) / num_positive
    
    return true_positive_rate


true_labels = test_labels
predicted_labels = predicted_label
positive_label = "Yes"  

confusion = confusion_matrix(true_labels, predicted_labels, positive_label)
precision_score = precision(confusion)
recall_score = recall(confusion)
f1_score_value = f1_score(precision_score, recall_score)
roc_auc_score_value = roc_auc_score(true_labels, predicted_labels)

# Print the results
print("Confusion Matrix:")
print("True Positives:", confusion["TP"])
print("True Negatives:", confusion["TN"])
print("False Positives:", confusion["FP"])
print("False Negatives:", confusion["FN"])
print("Precision:", precision_score)
print("Recall:", recall_score)
print("F1 Score:", f1_score_value)
print("ROC AUC Score:", roc_auc_score_value)
```
